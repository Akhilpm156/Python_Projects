```python
# Required lib imported
import requests as rq
import pandas as pd
from bs4 import BeautifulSoup as bs
```

## Begin to Scrap

```python
# Empty list to Store data
Name=[]
Price=[]
Review=[]
avali=[]

# for loop for get multiple pages data
for i in range(1,51):
    url=f"https://books.toscrape.com/catalogue/page-{i}.html"
    
    res=rq.get(url)

    sp=bs(res.text,"lxml")


    for i in sp.find_all('h3'):
        n=i.text
        Name.append(n)



    for i in sp.find_all('p',class_='price_color'):
        p=i.text[1:]
        Price.append(p)


    for i in sp.find_all('article'):
        Review.append(i.p['class'][1])


    for i in sp.find_all('p',class_='availability'):
        n=i.text.strip()
        avali.append(n)



data={'Product Name':Name,'Product Price':Price,'Product Review':Review,'Product Status':avali}
fdata=pd.DataFrame(data)

# Final Data Stored in csv formate

fdata.to_csv('final data.csv',index=False)
fdata
```
